### yolov8n_1 تقييم المودل

```python
from ultralytics import YOLO
import pandas as pd

# 1. تحميل النموذج
model = YOLO(r'E:\مجلد جديد\Receipt\runs\detect\yolov8n_1\weights\best.pt')

# 2. تعريف المجموعات التي نريد تقييمها
splits = ['train', 'val', 'test']
results_list = []

print("بدء عملية التقييم التفصيلية...")

for split in splits:
    print(f"جاري تقييم مجموعة: {split}...")
    
    metrics = model.val(split=split, plots=False)
    
    # استخراج المقاييس المطلوبة
    results_list.append({
        'Split': split,
        'Precision (P)': metrics.results_dict.get('metrics/precision(B)', 0),
        'Recall (R)': metrics.results_dict.get('metrics/recall(B)', 0),
        'mAP50': metrics.results_dict.get('metrics/mAP50(B)', 0), 
        'mAP50-95': metrics.results_dict.get('metrics/mAP50-95(B)', 0)
    })

# 3. عرض النتائجق
df = pd.DataFrame(results_list)
print("\n--- ملخص النتائج النهائي ---")
print(df.to_string(index=False))
```

# التنبا باستخدام المودل وحفظ الصور الفرعيه

```python
from ultralytics import YOLO
import os
import cv2  

model_path = r'E:\Senior\Receipt_Recognition\Receipt_Object_Detection\runs\detect\yolov8m_32\weights\best.pt'
source_img = r'E:\Senior\Receipt_Recognition\Receipt_Object_Detection\‪‪Invoice 21.jpg'

if os.path.exists(model_path) and os.path.exists(source_img):
    model = YOLO(model_path)
    results = model.predict(source=source_img, save=False, conf=0.35)
    
    original_image = cv2.imread(source_img)

    for i, result in enumerate(results):
        boxes = result.boxes.xyxy.cpu().numpy()  
        
        for j, box in enumerate(boxes):
            x1, y1, x2, y2 = map(int, box)
            
            cropped_part = original_image[y1:y2, x1:x2]
            
            crop_name = f'crop_{i}_{j}.jpg'
            cv2.imwrite(crop_name, cropped_part)
            
            print(f"✅ تم حفظ الجزء المقصوص: {crop_name}")
            
else:
    print("❌ تأكد من وجود الموديل والصورة في المسارات الصحيحة.")
```

#  بث حي للكاميرا

```python
import cv2
import torch
from ultralytics import YOLO
from IPython.display import display
import ipywidgets as widgets
import time

model_path = r'E:\Senior\Receipt_Recognition\Receipt_Object_Detection\runs\detect\yolov8m_32\weights\best.pt'

print("جاري تحميل النموذج على الـ GPU...")

model = YOLO(model_path).to('cuda') 
print("تم التحميل بنجاح.")

camera_running = True

def stop_camera_callback(b):
    global camera_running
    camera_running = False
    print("إيقاف الكاميرا... يرجى الانتظار")

stop_button = widgets.Button(description="إيقاف البث الحي", button_style='danger')
stop_button.on_click(stop_camera_callback)

image_widget = widgets.Image(format='jpeg', width=640, height=480)

display(stop_button)
display(image_widget)

cap = cv2.VideoCapture(0, cv2.CAP_DSHOW)

if not cap.isOpened():
    print("خطأ: لم نتمكن من فتح الكاميرا. تأكد من الصلاحيات أو أن الكاميرا ليست قيد الاستخدام.")
    camera_running = False

fps_start_time = time.time()
fps_counter = 0

try:
    while camera_running:
        ret, frame = cap.read()
        if not ret:
            break
        
        results = model.predict(source=frame, imgsz=320, conf=0.4, half=True, verbose=False)
        
        annotated_frame = results[0].plot()
        
        fps_counter += 1
        elapsed_time = time.time() - fps_start_time
        if elapsed_time > 1.0:
            current_fps = fps_counter / elapsed_time
            cv2.putText(annotated_frame, f"FPS: {current_fps:.1f}", (20, 40), 
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
            fps_counter = 0
            fps_start_time = time.time()

        _, buffer = cv2.imencode('.jpg', annotated_frame)
        image_widget.value = buffer.tobytes()

except Exception as e:
    print(f"حدث خطأ غير متوقع: {e}")

finally:
    cap.release()
    image_widget.close()
    print("تم تحرير الكاميرا والموارد بنجاح.")
```

```python
import cv2
import torch
from ultralytics import YOLO
import time

=model_path = r'E:\Senior\Receipt_Recognition\Receipt_Object_Detection\runs\detect\yolov8m_32\weights\best.pt'

print("جاري تحميل النموذج ونقله إلى الـ GPU...")
try:
    model = YOLO(model_path).to('cuda')
    print("تم تحميل النموذج بنجاح.")
except Exception as e:
    print(f"خطأ في تحميل النموذج: {e}")
    exit()

cap = cv2.VideoCapture(0, cv2.CAP_DSHOW)

window_name = 'Venezia Bank - Receipt Recognition Live'
cv2.namedWindow(window_name, cv2.WINDOW_NORMAL)

cv2.resizeWindow(window_name, 1280, 720)

if not cap.isOpened():
    print("خطأ: تعذر الوصول للكاميرا.")
else:
    print("تم فتح نافذة البث. يمكنك تكبيرها بالماوس الآن.")
    print("نصيحة: اضغط 'f' لملء الشاشة، و 'q' للإغلاق.")

fps_start_time = time.time()
fps_counter = 0
is_fullscreen = False

try:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret: break

        results = model.predict(source=frame, imgsz=320, conf=0.4, half=True, verbose=False)
        
        annotated_frame = results[0].plot()

        fps_counter += 1
        elapsed_time = time.time() - fps_start_time
        if elapsed_time > 1.0:
            current_fps = fps_counter / elapsed_time
            cv2.putText(annotated_frame, f"FPS: {current_fps:.1f}", (20, 50), 
                        cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 255, 0), 2)
            fps_counter = 0
            fps_start_time = time.time()

        cv2.imshow(window_name, annotated_frame)

        key = cv2.waitKey(1) & 0xFF
        
        if key == ord('q'):
            break
        
        elif key == ord('f'):
            is_fullscreen = not is_fullscreen
            if is_fullscreen:
                cv2.setWindowProperty(window_name, cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)
            else:
                cv2.setWindowProperty(window_name, cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_NORMAL)

except Exception as e:
    print(f"حدث خطأ: {e}")

finally:
    cap.release()
    cv2.destroyAllWindows()
    print("تم إغلاق البرنامج بنجاح.")
```

# التنبا بالصوره ورسم صناديق عليها

```python
from ultralytics import YOLO
import os
import cv2

# 1. إعداد المسارات 

model_path = r'E:\Senior\Receipt_Recognition\Receipt_Object_Detection\runs\detect\yolov8m_32\weights\best.pt'
source_img = r'E:\Senior\Receipt_Recognition\Receipt_Object_Detection\Invoice 11.webp'
output_dir = 'Prediction_Results'

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 2. التحقق من وجود الملفات وتحميل الموديل
if os.path.exists(model_path) and os.path.exists(source_img):
    print("🚀 جاري تحميل الموديل وبدء التنبؤ...")
    model = YOLO(model_path)
    
    # 3. عملية التنبؤ (Prediction)
    results = model.predict(source=source_img, save=False, conf=0.35, device='0') # استخدم device='cpu' إذا لم يكن لديك GPU
    
    # 4. رسم وتصوير التنبؤات على الصورة (Visualization)
    print("🎨 جاري رسم صناديق التنبؤ على الصورة...")
    
    annotated_image = results[0].plot() 

    # 5. حفظ الصورة الناتجة 
    base_name = os.path.basename(source_img)
    save_path = os.path.join(output_dir, f'predicted_{base_name}')
    
    # حفظ الصورة باستخدام OpenCV
    cv2.imwrite(save_path, annotated_image)
    
    print(f"✅ تم حفظ الصورة مع صناديق التنبؤ في: {save_path}")
    print("🔄 يمكنك الآن استخدام هذه الصورة في بحثك.")

else:
    print("❌ خطأ: تأكد من أن مسار الموديل ومسار الصورة صحيحان تماماً.")
    print(f"مسار الموديل المفترض: {model_path}")
    print(f"مسار الصورة المفترض: {source_img}")
```
